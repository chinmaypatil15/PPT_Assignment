```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.bottom = None


def merge(a, b):
    if a is None:
        return b
    if b is None:
        return a

    result = None

    if a.data < b.data:
        result = a
        result.bottom = merge(a.bottom, b)
    else:
        result = b
        result.bottom = merge(a, b.bottom)

    result.next = None
    return result


def flatten(root):
    if root is None or root.next is None:
        return root

    head = root
    while head.next:
        head.next = merge(head.next, head.next.next)
        head = head.next

    return root


def printList(head):
    temp = head
    while temp:
        print(temp.data, end=" ")
        temp = temp.bottom
    print()


# Test case 1
head = Node(5)
head.next = Node(10)
head.next.next = Node(19)
head.next.next.next = Node(28)

head.bottom = Node(7)
head.bottom.bottom = Node(8)
head.bottom.bottom.bottom = Node(30)

head.next.bottom = Node(20)

head.next.next.bottom = Node(22)
head.next.next.next.bottom = Node(35)
head.next.next.next.bottom.bottom = Node(40)
head.next.next.next.bottom.bottom.bottom = Node(45)

print("Input:")
printList(head)

head = flatten(head)

print("Output:")
printList(head)

# Test case 2
head = Node(5)
head.next = Node(10)
head.next.next = Node(19)
head.next.next.next = Node(28)

head.bottom = Node(7)
head.bottom.bottom = Node(22)

head.next.bottom = Node(8)

head.next.next.bottom = Node(50)

head.next.bottom.bottom = Node(30)

print("\nInput:")
printList(head)

head = flatten(head)

print("Output:")
printList(head)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def rotateLeft(self, head, k):
        if not head or not head.next or k == 0:
            return head
        length = 0
        curr = head
        while curr:
            length += 1
            curr = curr.next
        k = k % length

        if k == 0:
            return head
        curr = head
        for _ in range(k):
            curr = curr.next
        new_head = curr.next
        curr.next = None
        curr = new_head
        while curr.next:
            curr = curr.next
        curr.next = head
        return new_head


#case 1
N = 5
values = [2, 4, 7, 8, 9]
k = 3

# Create the linked list
head = None
for val in reversed(values):
    head = ListNode(val, head)

print("Input:")
curr = head
while curr:
    print(curr.val, end=" ")
    curr = curr.next

solution = Solution()
new_head = solution.rotateLeft(head, k)

print("\nOutput:")
curr = new_head
while curr:
    print(curr.val, end=" ")
    curr = curr.next


#case 2
N = 8
values = [1, 2, 3, 4, 5, 6, 7, 8]
k = 4

head = None
for val in reversed(values):
    head = ListNode(val, head)

print("\n\nInput:")
curr = head
while curr:
    print(curr.val, end=" ")
    curr = curr.next

solution = Solution()
new_head = solution.rotateLeft(head, k)

print("\nOutput:")
curr = new_head
while curr:
    print(curr.val, end=" ")
    curr = curr.next
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def nextLargerNodes(self, head):
        values = []
        curr = head
        while curr:
            values.append(curr.val)
            curr = curr.next
        answer = [0] * len(values)
        stack = []
        for i in range(len(values)-1, -1, -1):
            while stack and values[i] >= stack[-1]:
                stack.pop()
            if stack:
                answer[i] = stack[-1]
            stack.append(values[i])
        return answer

#case 1
head = ListNode(2)
head.next = ListNode(1)
head.next.next = ListNode(5)

print("Input: head = [2,1,5]")
solution = Solution()
result = solution.nextLargerNodes(head)
print("Output:", result)


#case 2
head = ListNode(2)
head.next = ListNode(7)
head.next.next = ListNode(4)
head.next.next.next = ListNode(3)
head.next.next.next.next = ListNode(5)

print("\nInput: head = [2,7,4,3,5]")
solution = Solution()
result = solution.nextLargerNodes(head)
print("Output:", result)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def removeZeroSumSublists(self, head):
        dummy = ListNode(0)
        dummy.next = head
        curr_sum = 0
        sum_map = {}
        sum_map[0] = dummy
        curr = dummy.next
        while curr:
            curr_sum += curr.val
            if curr_sum in sum_map:
                prev = sum_map[curr_sum].next
                remove_sum = curr_sum + prev.val
                while prev != curr:
                    sum_map.pop(remove_sum)
                    prev = prev.next
                    remove_sum += prev.val
                sum_map[curr_sum].next = curr.next
            else:
                sum_map[curr_sum] = curr
            curr = curr.next
        return dummy.next


# Test case 1
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(-3)
head.next.next.next = ListNode(3)
head.next.next.next.next = ListNode(1)

print("Input: head = [1,2,-3,3,1]")
solution = Solution()
result = solution.removeZeroSumSublists(head)
output = []
while result:
    output.append(result.val)
    result = result.next
print("Output:", output)


# Test case 2
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(-3)
head.next.next.next.next = ListNode(4)

print("\nInput: head = [1,2,3,-3,4]")
solution = Solution()
result = solution.removeZeroSumSublists(head)
output = []
while result:
    output.append(result.val)
    result = result.next
print("Output:", output)


# Test case 3
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(-3)
head.next.next.next.next = ListNode(-2)

print("\nInput: head = [1,2,3,-3,-2]")
solution = Solution()
result = solution.removeZeroSumSublists(head)
output = []
while result:
    output.append(result.val)
    result = result.next
print("Output:", output)
```
